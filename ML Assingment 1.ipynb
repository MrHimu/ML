```python
#The Data
#Name : The Complete Pokemon Dataset
#Source : Kaggle
"""Context

This dataset contains information on all 802 Pokemon from all Seven Generations of Pokemon. The information contained in this dataset include Base Stats, Performance against Other Types, Height, Weight, Classification, Egg Steps, Experience Points, Abilities, etc. The information was scraped from http://serebii.net/
Content

    name: The English name of the Pokemon
    japanese_name: The Original Japanese name of the Pokemon
    pokedex_number: The entry number of the Pokemon in the National Pokedex
    percentage_male: The percentage of the species that are male. Blank if the Pokemon is genderless.
    type1: The Primary Type of the Pokemon
    type2: The Secondary Type of the Pokemon
    classification: The Classification of the Pokemon as described by the Sun and Moon Pokedex
    height_m: Height of the Pokemon in metres
    weight_kg: The Weight of the Pokemon in kilograms
    capture_rate: Capture Rate of the Pokemon
    base_egg_steps: The number of steps required to hatch an egg of the Pokemon
    abilities: A stringified list of abilities that the Pokemon is capable of having
    experience_growth: The Experience Growth of the Pokemon
    base_happiness: Base Happiness of the Pokemon
    against_?: Eighteen features that denote the amount of damage taken against an attack of a particular type
    hp: The Base HP of the Pokemon
    attack: The Base Attack of the Pokemon
    defense: The Base Defense of the Pokemon
    sp_attack: The Base Special Attack of the Pokemon
    sp_defense: The Base Special Defense of the Pokemon
    speed: The Base Speed of the Pokemon
    generation: The numbered generation which the Pokemon was first introduced
    is_legendary: Denotes if the Pokemon is legendary.

Acknowledgements

The data was scraped from http://serebii.net/.
Inspiration

Pokemon holds a very special place in my heart as it is probably the only video game I have judiciously followed for more than 10 years. With this dataset, I wanted to be able to answer the following questions:

    Is it possible to build a classifier to identify legendary Pokemon?
    How does height and weight of a Pokemon correlate with its various base stats?
    What factors influence the Experience Growth and Egg Steps? Are these quantities correlated?
    Which type is the strongest overall? Which is the weakest?
    Which type is the most likely to be a legendary Pokemon?
    Can you build a Pokemon dream team? A team of 6 Pokemon that inflicts the most damage while remaining relatively impervious to any other team of 6 Pokemon."""
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pokemon_data = pd.read_csv('pokemon.csv')
```

```python
# Univariate Analysis
# Descriptive statistics for numerical columns
numerical_stats = pokemon_data.describe()
print("Summary statistics for numerical variables:\n", numerical_stats)
```

```python

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Attack distribution
sns.histplot(pokemon_data['attack'], kde=True, ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Attack Distribution')

# Defense distribution
sns.histplot(pokemon_data['defense'], kde=True, ax=axes[0, 1], color='salmon')
axes[0, 1].set_title('Defense Distribution')

# Speed distribution
sns.histplot(pokemon_data['speed'], kde=True, ax=axes[1, 0], color='lightgreen')
axes[1, 0].set_title('Speed Distribution')

# Weight distribution
sns.histplot(pokemon_data['weight_kg'], kde=True, ax=axes[1, 1], color='gold')
axes[1, 1].set_title('Weight Distribution')
plt.tight_layout()
plt.show()
```

```python
# Visualizing Categorical variable (Type1)
plt.figure(figsize=(10, 6))
sns.countplot(x='type1', data=pokemon_data, palette='Set2', order=pokemon_data['type1'].value_counts().index)
plt.title('Distribution of Pokémon Types')
plt.xticks(rotation=45)
plt.show()
```

```python
# Boxplots for visualizing distributions and outliers
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Boxplot for Attack
sns.boxplot(data=pokemon_data, x='attack', ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Attack Boxplot')

# Boxplot for Defense
sns.boxplot(data=pokemon_data, x='defense', ax=axes[0, 1], color='salmon')
axes[0, 1].set_title('Defense Boxplot')

# Boxplot for HP
sns.boxplot(data=pokemon_data, x='hp', ax=axes[1, 0], color='lightgreen')
axes[1, 0].set_title('HP Boxplot')

# Boxplot for Speed
sns.boxplot(data=pokemon_data, x='speed', ax=axes[1, 1], color='gold')
axes[1, 1].set_title('Speed Boxplot')

plt.tight_layout()
plt.show()





```

```python
# Bar plot for a categorical variable: Pokémon Generation
plt.figure(figsize=(10, 6))
sns.countplot(x='generation', data=pokemon_data)
plt.title('Distribution of Pokémon Generations')
plt.show()
```

```python
# Plotting the percentage of legendary Pokémon
legendary_counts = pokemon_data['is_legendary'].value_counts(normalize=True) * 100
legendary_labels = ['Non-Legendary', 'Legendary']

plt.figure(figsize=(7, 7))
plt.pie(legendary_counts, labels=legendary_labels, autopct='%1.1f%%', colors=['lightblue', 'lightcoral'], startangle=90)
plt.title('Percentage of Legendary vs Non-Legendary Pokémon')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.show()
```

```python
#  Boxplot for weight categorized by Type
plt.figure(figsize=(12, 6))
sns.boxplot(x='type1', y='weight_kg', data=pokemon_data, palette='Set3')
plt.xticks(rotation=45)
plt.title('Weight Distribution by Pokémon Type')
plt.show()
```

```python
#**Multivariate**
```

```python
# Correlation matrix for numerical variables
correlation_matrix = pokemon_data[['attack', 'defense', 'sp_attack', 'sp_defense', 'speed', 'hp', 'weight_kg']].corr()

# Plot the correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix for Key Attributes')
plt.show()


```

```python
# Pairplot for multivariate relationships between numerical attributes
sns.pairplot(pokemon_data[['attack', 'defense', 'sp_attack', 'sp_defense', 'speed']])
plt.show()
```

```python
#  Heatmap of correlations between key attributes for specific Pokémon generations
plt.figure(figsize=(10, 8))
# Generating correlation matrix for Generation 1 Pokémon only
gen1_data = pokemon_data[pokemon_data['generation'] == 1]
gen1_corr = gen1_data[['attack', 'defense', 'sp_attack', 'sp_defense', 'speed', 'hp', 'weight_kg']].corr()

sns.heatmap(gen1_corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix for Generation 1 Pokémon')
plt.show()


```

```python
#  Grouped bar plot: Mean stats by Pokémon type (Filter numerical columns)
# Ensure that only numerical columns are selected
numerical_columns = ['attack', 'defense', 'speed']
mean_stats_by_type = pokemon_data.groupby('type1')[numerical_columns].mean()
```

```python
# Scatter plot with regression lines to explore relationships
# Attack vs. Speed with regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='attack', y='speed', data=pokemon_data, scatter_kws={'s':20}, line_kws={'color':'red'})
plt.title('Attack vs Speed with Regression Line')
plt.show()

```

```python
# Sp. Attack vs Sp. Defense with regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='sp_attack', y='sp_defense', data=pokemon_data, scatter_kws={'s':20}, line_kws={'color':'blue'})
plt.title('Sp. Attack vs Sp. Defense with Regression Line')
plt.show()

```

```python
#  Pairplot categorized by Generation (adding hue to show distinction between Generations)
sns.pairplot(pokemon_data[['attack', 'defense', 'sp_attack', 'sp_defense', 'speed', 'generation']], hue='generation', palette='tab10')
plt.title('Pairplot of Key Attributes Categorized by Generation')
plt.show()

```

```python
#  Violin plot to show the distribution of Speed by Pokémon Type
plt.figure(figsize=(12, 6))
sns.violinplot(x='type1', y='speed', data=pokemon_data, palette='Set2')
plt.xticks(rotation=45)
plt.title('Speed Distribution by Pokémon Type')
plt.show()

```

```python
#  Heatmap showing Pokémon count by Generation and Legendary Status
plt.figure(figsize=(8, 6))
legendary_heatmap = pokemon_data.pivot_table(index='generation', columns='is_legendary', aggfunc='size')
sns.heatmap(legendary_heatmap, annot=True, cmap='YlGnBu', linewidths=0.5)
plt.title('Heatmap of Pokémon Count by Generation and Legendary Status')
plt.show()
```
